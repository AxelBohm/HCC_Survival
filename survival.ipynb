```python
import pandas as pd
import numpy as np
#import matplotlib.pyplot as plt
```

```python
df_missing = pd.read_csv('./hcc-survival/hcc-data.txt')
```

```python
df_missing.replace("?", np.NaN, inplace=True)
df_missing.head()
```

```python
df = df_missing.astype('float')
df.fillna(df.mean(), inplace=True)
df.head()
```

```python
df.corr
```

```python
df.describe()
```

```python
X = df.values[:,0:49]
y = df.values[:,49]
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

# evaluate an LDA model on the dataset using k-fold cross validation
model = LinearDiscriminantAnalysis()
kfold = KFold(n_splits=10)
result = cross_val_score(model, X, y, cv=kfold, scoring='accuracy')
print("LDA: ", result.mean())
```

```python
from sklearn import svm
clf = svm.SVC()
kfold_svm = KFold(n_splits=10)
result_svm = cross_val_score(clf, X, y, cv=kfold_svm)
print("svc: ", result_svm.mean())
```

```python
model = svm.NuSVC()
res = cross_val_score(model, X, y, cv=kfold)
print("NuSVC: ", res.mean())
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier()
res = cross_val_score(forest, X, y, cv=kfold) 
print("RandomForest: ", res.mean())
```

```python
forest.fit(X, y)
```

```python
feature_importance = dict(zip(df.columns.values,forest.feature_importances_))
```

```python
sorted(feature_importance.items(), key=lambda kv: kv[1], reverse=True)
```

## XGBoost

```python
import xgboost as xgb
X_m = df_missing.values[:,0:49]
y = df_missing.values[:,49]
xg_train = xgb.DMatrix(X, label=y, missing=-999.0)

param = {'max_depth': 2, 'eta': 1, 'silent': 1, 'objective': 'binary:logistic'}
param['nthread'] = 4
param['eval_metric'] = 'error'
param['reg_alpha'] = 0.1
#param['reg_lambda'] = 1
param['rate_drop'] = 0.1

n_folds = 10
cv = xgb.cv(param, xg_train, 100, nfold=n_folds, verbose_eval=1)
```

```python
param_boost_linear = param
param_boost_linear['booster'] = 'gblinear'
n_folds = 5
cv = xgb.cv(param_boost_linear, xg_train, 100, nfold=n_folds, verbose_eval=1)
```

```python
num_round = 10
bst = xgb.train(param, xg_train, num_round)
print(bst.feature_names)
```
